# Masking Email Addresses Function

I'll create a function to mask email addresses by hiding part of the username while preserving the domain. This is useful for privacy protection when displaying email addresses.

## Function Implementation


```python
import re

def mask_email(email):
    """
    Masks an email address by replacing part of the username with asterisks.
    
    Args:
        email (str): The email address to mask
        
    Returns:
        str: The masked email address
    
    Examples:
        >>> mask_email("john.doe@example.com")
        'j****oe@example.com'
        >>> mask_email("alice@gmail.com")
        'a***e@gmail.com'
    """
    if not email or '@' not in email:
        return "Invalid email format"
    
    # Split the email into username and domain
    username, domain = email.split('@', 1)
    
    # Handle different username lengths
    if len(username) <= 2:
        masked_username = username[0] + '*' * (len(username) - 1)
    else:
        # Keep first and last character, mask the rest
        masked_username = username[0] + '*' * (len(username) - 2) + username[-1]
    
    return f"{masked_username}@{domain}"

# Examples
emails = [
    "john.doe@example.com",
    "alice@gmail.com",
    "bob.smith123@company.co.uk",
    "a@b.com",
    "verylongemail.address@domain.org"
]

for email in emails:
    print(f"Original: {email}")
    print(f"Masked:   {mask_email(email)}")
    print()
```


This function:
1. Splits the email at the '@' symbol
2. Preserves the first and last characters of the username
3. Replaces the middle characters with asterisks
4. Keeps the domain intact
5. Handles edge cases like very short usernames
